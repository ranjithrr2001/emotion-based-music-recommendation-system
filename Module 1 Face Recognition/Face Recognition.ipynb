## Haar Cascade Classifier

```python
import cv2

data = "../Datasets/"

pixels = cv2.imread(data+'happy1.jpeg')
classifier = cv2.CascadeClassifier(data+'haarcascade_frontalface_default.xml')
faces = classifier.detectMultiScale(pixels)
for box in faces:
	x, y, width, height = box
	x2, y2 = x + width, y + height
	cv2.rectangle(pixels, (x, y), (x2, y2), (0,0,255), 1)
cv2.imshow('face detection', pixels)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

data = "../Datasets/"
classifier = cv2.CascadeClassifier(data+'haarcascade_frontalface_default.xml')
cap = cv2.VideoCapture(0)
while True:
    try:
        _, frame = cap.read()
        faces = classifier.detectMultiScale(frame)
        for box in faces:
            x, y, w, h = box
            x1, y1 = x+w, y+h
            cv2.rectangle(frame,(x,y),(x1,y1),(0,0,255))
        cv2.imshow('Haar Cascade Classifier',frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    except Exception as e:
        print(e)
        break
cap.release()
cv2.destroyAllWindows()
```

## MTCNN - Multi Task Cascaded Convolutional Neural Network 

```python
import cv2
from mtcnn.mtcnn import MTCNN

detector = MTCNN()
pixels = cv2.imread(data+'happy1.jpeg')
faces = detector.detect_faces(pixels)
for box in faces:
	x, y, width, height = box['box']
	x2, y2 = x + width, y + height
	cv2.rectangle(pixels, (x, y), (x2, y2), (0,0,255), 1)
cv2.imshow('face detection', pixels)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
